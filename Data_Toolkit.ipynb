#Q.3:- Explain the following terms:
1. The difference in np.array, np.asarray and np.asanyarray.
2. The difference between Deep copy and shallow copy.

1. Difference between np.array, np.asarray, and np.asanyarray
np.array:

**Purpose:** Creates a new array from any object that exposes the array interface.

**Behavior**: Always creates a new array, even if the input is already an array.

**Usage:** Use when you need a new array and want to ensure a copy is made.

**Example:**

```python
import numpy as np
a = [1, 2, 3]
b = np.array(a)
```

**np.asarray:**

**Purpose:** Converts the input to an array.

**Behavior:** Does not create a copy if the input is already an array with the same dtype.

**Usage:** Use when you want to avoid unnecessary copying and the input might already be an array.

**Example :**

```python
import numpy as np
a = [1, 2, 3]
b = np.asarray(a)
```

**np.asanyarray:**

**Purpose:** Converts the input to an array, but passes through subclasses of ndarray.

**Behavior:** Similar to np.asarray, but if the input is a subclass of ndarray, it returns the input as is.

**Usage:** Use when you want to preserve subclasses of ndarray.

**Example:**

```python
import numpy as np
a = np.matrix([[1, 2], [3, 4]])
b = np.asanyarray(a)
```

Difference between Deep Copy and Shallow Copy

**Shallow Copy:**

**Definition:** Creates a new object, but inserts references into it to the objects found in the original.

**Behavior:** Changes to the copied object may affect the original object if they involve mutable objects.

**Usage:** Use when you need a new object but want to share the same data.

**Example:**

```python
import copy
original = [[1, 2, 3], [4, 5, 6]]
shallow_copy = copy.copy(original)
shallow_copy[0][0] = 99
print(original)  # Output: [[99, 2, 3], [4, 5, 6]]
```

**Deep Copy:**

**Definition:** Creates a new object and recursively copies all objects found in the original.

**Behavior:** Changes to the copied object do not affect the original object.
Usage: Use when you need a completely independent copy of the original object.

**Example:**

```python
import copy
original = [[1, 2, 3], [4, 5, 6]]
deep_copy = copy.deepcopy(original)
deep_copy[0][0] = 99
print(original)  # Output: [[1, 2, 3], [4, 5, 6]]
```

```python
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Generate random data for temperature and humidity
np.random.seed(0)  # for reproducibility
temperature = np.random.uniform(-10,35, size=len(date_range))
humidity = np.random.uniform(20,90, size=len(date_range))

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot Temperature with the primary y-axis
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')
ax1.legend()

# Create a secondary y-axis for the Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.legend()
# Add a title
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# b) Label the x-axis as 'Date'.


import matplotlib.pyplot as plt

# Plot Temperature and Humidity over time
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Temperature'], label='Temperature (°C)', color='red')
plt.plot(df['Date'], df['Humidity'], label='Humidity (%)', color='blue')

# Label the axes
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Temperature and Humidity Over Time')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

#Q.15 :- Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value and the y-axis as Frequency/Probability.

d) Set the title of the plot as Histogram with PDF Overlay

```python
import numpy as np

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Plot the histogram of the data with 30 bins
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Overlay the PDF
mu, std = norm.fit(data)
pdf = norm.pdf(bins, mu, std)
plt.plot(bins, pdf, 'r--', linewidth=2, label='PDF')

# Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

#Q.17 :- Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrants
colors = ['blue' if (x_val > 0 and y_val > 0) else 'green' if (x_val < 0 and y_val > 0)
          else 'red' if (x_val < 0 and y_val < 0) else 'orange' for x_val, y_val in zip(x, y)]

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, hue=colors, palette=['blue', 'green', 'red', 'orange'], legend='full')

# Labeling the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Add legend for quadrants
quadrant_labels = {'blue': 'Q1 (x>0, y>0)', 'green': 'Q2 (x<0, y>0)', 'red': 'Q3 (x<0, y<0)', 'orange': 'Q4 (x>0, y<0)'}
handles = [plt.Line2D([0], [0], marker='o', color='w', label=label, markersize=10, markerfacecolor=color)
           for color, label in quadrant_labels.items()]
plt.legend(handles=handles, title='Quadrants')

# Display the plot
plt.show()
```

#Q.18:- With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable output in Jupyter Notebook
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the results
show(p)
```

#Q.19 :- Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Enable output in Jupyter Notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 101, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a new plot with a title and axis labels
p = figure(x_range=categories, title="Random Categorical Bar Chart", x_axis_label='Categories', y_axis_label='Values')

# Add bars to the plot
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color='blue')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show the results
show(p)
```
